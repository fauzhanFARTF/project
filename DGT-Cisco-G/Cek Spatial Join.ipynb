```python
bangunan_toko_modern = gpd.read_file(r'assets/Toko_Modern_Centroids.json')
```

```python
bangunan_toko_modern
```

```python
bangunan_toko_modern_selected = bangunan_toko_modern[['N_Alfamart','Desa','Kecamatan', 'NAMA_PERUS',  'geometry']]
bangunan_toko_modern_selected
```

```python
Kolombaru_minimarket = {'N_Alfamart':'Nama Minimarket', 'NAMA_PERUS':'Nama Perusahaan','Desa' : 'Nama Desa', 'Kecamatan':'Nama Kecamatan' } 
bangunan_toko_modern_selected.rename(columns=Kolombaru_minimarket, inplace=True)
bangunan_toko_modern_selected
```

```python
import folium
m = folium.Map(location=[-6.1950, 106.5528], zoom_start=11, tiles='CartoDB positron')

for _, r in bangunan_toko_modern_selected.iterrows():
    # Without simplifying the representation of each borough,
    # the map might not be displayed
    sim_geo = gpd.GeoSeries(r['geometry']).simplify(tolerance=0.001)
    geo_j = sim_geo.to_json()
    geo_j = folium.GeoJson(data=geo_j,
                           style_function=lambda x: {'fillColor': 'orange', 'fillOpacity' : 0, 'weight': 1})
    folium.Popup(r['Nama Minimarket']).add_to(geo_j)
    geo_j.add_to(m)
m
```
